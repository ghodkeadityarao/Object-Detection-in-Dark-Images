```python
#Preprocessing the annotations file by changing label name to 0 indexed numbers

os.chdir("D:\\Semester 6\\Deep Learning\\Project")
ROOT_DIR = os.getcwd()

classes = {}
with open("classes.txt", "r") as myFile:
    for num, line in enumerate(myFile, 0):
        line = line.rstrip("\n")
        classes[line] = num
    myFile.close()
    
os.chdir("ExDark_Annno")
DIRS = os.listdir(os.getcwd())

print(classes)
```

```python

for DIR in DIRS:

    if os.path.isdir(DIR):
        
        os.chdir(DIR)
        print("Currently in:", DIR)


        for filename in tqdm(os.listdir(os.getcwd())):
            filename_str = str.split(filename, ".")[0]

            if filename.endswith(".txt"):
                annotations = []

                with open(filename) as f:
                    for l,values in enumerate(f):
                        if l==0:
                            continue
                        classLabel = values.split()[0]
                        if classLabel.lower() not in list(classes.keys()):
                            continue
                            
                        values = values.replace(classLabel,classLabel.lower())
                        
                        values = values.replace(classLabel.lower(), str(classes[classLabel.lower()]))

                        label_value = values.split()
                        coords = np.asarray([float(label_value[1]), float(label_value[2]), float(label_value[3]), float(label_value[4])])
            
                        label_value[1], label_value[2], label_value[3], label_value[4] = coords[0], coords[1], coords[2], coords[3]
                        newvalues = str(label_value[0]) + " " + str(label_value[1]) + " " + str(label_value[2]) + " " + str(label_value[3]) + " " + str(label_value[4])
                        values = values.replace(values, newvalues)
                        annotations.append(values)
                    f.close()
               
                with open(filename, "w") as outfile:
                    for values in annotations:
                        outfile.write(values)
                        outfile.write("\n")
                    outfile.close()    

```

```python
import os
import shutil

def copy_images(image_paths_file, destination_folder):
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    with open(image_paths_file, 'r') as file:
        image_paths = file.readlines()

    for image_path in image_paths:
        image_path = image_path.strip() 
        if os.path.exists(image_path): 
            image_name = os.path.basename(image_path) 
            destination_path = os.path.join(destination_folder, image_name)
            shutil.copyfile(image_path, destination_path)
            print(f"Copied {image_name} to {destination_folder}")
        else:
            print(f"File {image_path} does not exist.")

# Example usage:
if __name__ == "__main__":
    image_paths_file = "D:/Semester 6/Deep Learning/Project/train_label.txt" 
    destination_folder = "D:/Semester 6/Deep Learning/Project/datasets/dataset_yolo/train/labels" 
    copy_images(image_paths_file, destination_folder)
```
